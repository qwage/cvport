# Demo of B-spline Representation of Node Trajectory 

```python
# importing necessary modules 
import numpy as np 
from scipy.interpolate import splprep, splev 
import matplotlib.pyplot as plt 
```

```python
# 2d Node Trajectory  

pts = [[-20.494, -43.698],
       [-27.710, -21.882],
       [34.595, -20.910],
       [28.018,  22.085],
       [11.885,   4.075]]


pts = np.asarray(pts, dtype=float)

# Or generate nodes randomly 
# pts = np.random.uniform(-50, 50, (5, 2))
```

```python
# Plot the original node trajectory 
plt.figure()
plt.scatter(pts[:, 0], pts[:, 1], c='r')
plt.plot(pts[:, 0], pts[:, 1], '-b')
plt.show()
```

```python
# Conduct spline interpolation of the node data 

# Try changing the parameter s to a different value besides 100
tck, u = splprep([pts[:, 0], pts[:, 1]], k=2, s=50)
print(u)
```

```python
# Creating new u-parameter points that we want the spline to pass through
# u-parameter MUST be 0 <= u <= 1

# Try adjusting the number of u_new points to understand how the B-spline works 
u_new = np.arange(0, 1.01, 0.05)
```

```python
# Generate new points 
x_new, y_new = splev(u_new, tck)
```

```python
# Plot the spline fitted node trajectory 
plt.figure()
plt.scatter(x_new, y_new, c='r')
plt.plot(x_new, y_new, '-b')
plt.scatter(pts[:, 0], pts[:, 1], c='r')
plt.plot(pts[:, 0], pts[:, 1], '-b')
plt.title('B-spline Trajectory, s=100')
plt.show()
```
